```python
import pandas as pd
```

dataset robot movement vertikal and horizotal with 2 motors

```python
datarobot=pd.read_csv("dataset4.csv")
print(datarobot)
```

```python
data_vertikal=pd.DataFrame(datarobot,columns=['sensor v1','motor v1'])
x_vertikal=data_vertikal.iloc[:,0:1].values #mengambil data input
y_vertikal=data_vertikal.iloc[:,-1].values #mengambil data output
print(x_vertikal)
print(y_vertikal)
```

```python
from sklearn.model_selection import train_test_split
x_train_vertikal, x_test_vertikal, y_train_vertikal, y_test_vertikal = train_test_split(x_vertikal, y_vertikal, test_size=0.2, random_state=0)
```

```python
# x_train = x_train.flatten()
y_test_vertikal = y_test_vertikal.flatten()
y_test_vertikal
```

```python
y_train_vertikal.shape
```

Training Datast and finding best parameter epoch and batch_size to get the best model

```python
import tensorflow as tf

model_vertikal=tf.keras.models.Sequential()

#model.add(tf.keras.layers.Dense(units=3, activation='relu'))
model_vertikal.add(tf.keras.layers.Dense(units=8, activation='relu'))
model_vertikal.add(tf.keras.layers.Dense(units=1))
model_vertikal.compile(loss='mean_absolute_error',optimizer=tf.keras.optimizers.Adam(0.004))

validasi = model_vertikal.fit(x_train_vertikal, y_train_vertikal, validation_data=(x_test_vertikal, y_test_vertikal), epochs=200,batch_size=8)
```

```python
results = model_vertikal.evaluate(x_test_vertikal, y_test_vertikal, batch_size=8)
print("test loss, test acc:", results)
```

```python
model_vertikal.save("bnn_vertikal_termaik1.h5")
```

```python
import matplotlib.pyplot as plt

print(model_vertikal.predict([[105]]))
```

Graphic of data prediction and data test

```python
plt.figure(figsize=(10, 4))
plt.plot(model_vertikal.predict(x_test_vertikal), label='Predict', color='red')
plt.plot(y_test_vertikal,label='Test', color='blue')
plt.ylabel("prediksi putaran stepper motor vertikal")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
#make ploting loss function MSE

plt.plot(validasi.history['loss'])
plt.plot(validasi.history['val_loss'], color="orange")
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train test', 'val test'], loc='upper left')
plt.show()
```

Horizontal

```python
datarobot1=pd.read_csv("dataset_hor.csv")
print(datarobot1)
```

```python
data_horizontal=pd.DataFrame(datarobot1,columns=['sensor h1','motor h1'])
x_horizontal=data_horizontal.iloc[:,0:1].values #mengambil data input
y_horizontal=data_horizontal.iloc[:,-1].values #mengambil data output
print(x_horizontal)
print(y_horizontal)
```

```python
from sklearn.model_selection import train_test_split
x_train_horizontal, x_test_horizontal, y_train_horizontal, y_test_horizontal = train_test_split(x_horizontal, y_horizontal, test_size=0.2, random_state=0)
```

```python
# x_train = x_train.flatten()
y_test_horizontal = y_test_horizontal.flatten()
y_test_horizontal
```

```python
import tensorflow as tf

model_horizontal=tf.keras.models.Sequential()

#model.add(tf.keras.layers.Dense(units=3, activation='relu'))
model_horizontal.add(tf.keras.layers.Dense(units=8, activation='relu'))
model_horizontal.add(tf.keras.layers.Dense(units=1))
model_horizontal.compile(loss='mean_absolute_error',optimizer=tf.keras.optimizers.Adam(0.005))

validasi = model_horizontal.fit(x_train_horizontal, y_train_horizontal, validation_data=(x_test_horizontal, y_test_horizontal), epochs=300,batch_size=8)
```

```python
results = model_horizontal.evaluate(x_test_horizontal, y_test_horizontal, batch_size=8)
print("test loss, test acc:", results)
```

```python
#make ploting loss function MSE

plt.figure(figsize=(10, 4))
plt.plot(model_horizontal.predict(x_test_horizontal), label='Predict', color='red')
plt.plot(y_test_horizontal,label='Test', color='blue')
plt.ylabel("prediksi putaran stepper motor horizontal")
plt.xticks()
plt.yticks()
plt.legend()
plt.show()
```

```python
plt.plot(validasi.history['loss'])
plt.plot(validasi.history['val_loss'], color="orange")
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('learning rate')
plt.legend(['train test', 'val test'], loc='upper left')
plt.show()
```

Best model saved in .h5 file

```python
model_horizontal.save("bnn_horizontal_terbaik1.h5")
```
